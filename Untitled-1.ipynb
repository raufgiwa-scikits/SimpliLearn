```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, LSTM, Add, Concatenate
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Generate synthetic data
def generate_data(num_samples, time_steps, num_time_features, num_non_time_features):
    time_series_data = np.random.rand(num_samples, time_steps, num_time_features)  # Time-series data
    non_time_series_data = np.random.rand(num_samples, num_non_time_features)      # Non-time-series data
    targets = np.random.randint(0, 2, size=(num_samples, 1))                       # Binary classification
    return time_series_data, non_time_series_data, targets

# Parameters
num_samples = 1000
time_steps = 10
num_time_features = 3
num_non_time_features = 5

# Generate data
time_series_data, non_time_series_data, targets = generate_data(
    num_samples, time_steps, num_time_features, num_non_time_features
)

# Split the data into training and testing sets
X_time_train, X_time_test, X_non_time_train, X_non_time_test, y_train, y_test = train_test_split(
    time_series_data, non_time_series_data, targets, test_size=0.2, random_state=42
)

# Standardize non-time-series data
scaler = StandardScaler()
X_non_time_train = scaler.fit_transform(X_non_time_train)
X_non_time_test = scaler.transform(X_non_time_test)

# Define single-effect contribution model
def build_single_effect_branch(input_shape, feature_type, num_features):
    inputs = Input(shape=input_shape, name=f"{feature_type}_Input")
    outputs = []
    for i in range(num_features):
        if feature_type == "time":
            feature = inputs[:, :, i:i+1]
            x = LSTM(16, activation='tanh', return_sequences=False)(feature)
        else:
            feature = inputs[:, i:i+1]
            x = Dense(16, activation='relu')(feature)
        x = Dense(1, activation='linear')(x)
        outputs.append(x)
    output_sum = Add()(outputs)
    return inputs, outputs, output_sum

# Define pairwise-effect contribution model
def build_pairwise_effect_branch(input_shape, feature_type, num_features):
    inputs = Input(shape=input_shape, name=f"{feature_type}_Input")
    outputs = []
    for i in range(num_features):
        for j in range(i + 1, num_features):
            if feature_type == "time":
                pair = Concatenate(axis=-1)([inputs[:, :, i:i+1], inputs[:, :, j:j+1]])
                x = LSTM(16, activation='tanh', return_sequences=False)(pair)
            else:
                pair = Concatenate(axis=-1)([inputs[:, i:i+1], inputs[:, j:j+1]])
                x = Dense(16, activation='relu')(pair)
            x = Dense(1, activation='linear')(x)
            outputs.append(x)
    output_sum = Add()(outputs)
    return inputs, outputs, output_sum

# Build single and pairwise models
time_single_input, time_single_outputs, time_single_sum = build_single_effect_branch(
    (time_steps, num_time_features), "time", num_time_features
)
non_time_single_input, non_time_single_outputs, non_time_single_sum = build_single_effect_branch(
    (num_non_time_features,), "non_time", num_non_time_features
)

time_pairwise_input, time_pairwise_outputs, time_pairwise_sum = build_pairwise_effect_branch(
    (time_steps, num_time_features), "time", num_time_features
)
non_time_pairwise_input, non_time_pairwise_outputs, non_time_pairwise_sum = build_pairwise_effect_branch(
    (num_non_time_features,), "non_time", num_non_time_features
)

# Combined final output for fine-tuning
final_output = Add()([time_single_sum, non_time_single_sum, time_pairwise_sum, non_time_pairwise_sum])
final_output = Dense(1, activation='sigmoid', name="Output")(final_output)

# Create the full model
model = Model(inputs=[time_single_input, non_time_single_input, time_pairwise_input, non_time_pairwise_input], outputs=final_output)

# Training functions
def train_single_effect(time_data, non_time_data, labels, optimizer, epochs=10, batch_size=32):
    for epoch in range(epochs):
        for i in range(0, len(time_data), batch_size):
            time_batch = time_data[i:i+batch_size]
            non_time_batch = non_time_data[i:i+batch_size]
            y_batch = labels[i:i+batch_size]
            with tf.GradientTape() as tape:
                y_pred = model([time_batch, non_time_batch], training=True)
                loss = tf.reduce_mean(tf.keras.losses.binary_crossentropy(y_batch, y_pred))
            gradients = tape.gradient(loss, model.trainable_variables)
            optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        print(f"Single Effect Epoch {epoch+1}/{epochs} Loss: {loss.numpy()}")

# Train pairwise effects

def train_pairwise_effect(time_data, non_time_data, labels, optimizer, epochs=10, batch_size=32):
    for epoch in range(epochs):
        for i in range(0, len(time_data), batch_size):
            time_batch = time_data[i:i + batch_size]
            non_time_batch = non_time_data[i:i + batch_size]
            y_batch = labels[i:i + batch_size]
            with tf.GradientTape() as tape:
                y_pred = model([time_batch, non_time_batch, time_batch, non_time_batch], training=True)
                loss = tf.reduce_mean(tf.keras.losses.binary_crossentropy(y_batch, y_pred))
            gradients = tape.gradient(loss, model.trainable_variables)
            optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        print(f"Pairwise Effect Epoch {epoch+1}/{epochs} Loss: {loss.numpy()}")

# Fine-tuning both single and pairwise effects together
def fine_tune_model(time_data, non_time_data, labels, optimizer, epochs=10, batch_size=32):
    for epoch in range(epochs):
        for i in range(0, len(time_data), batch_size):
            time_batch = time_data[i:i + batch_size]
            non_time_batch = non_time_data[i:i + batch_size]
            y_batch = labels[i:i + batch_size]
            with tf.GradientTape() as tape:
                y_pred = model([time_batch, non_time_batch, time_batch, non_time_batch], training=True)
                loss = tf.reduce_mean(tf.keras.losses.binary_crossentropy(y_batch, y_pred))
            gradients = tape.gradient(loss, model.trainable_variables)
            optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        print(f"Fine-Tune Epoch {epoch+1}/{epochs} Loss: {loss.numpy()}")

# Feature contributions for single and pairwise effects
def calculate_feature_contributions(time_data, non_time_data):
    time_single_contributions = [output(time_data).numpy() for output in time_single_outputs]
    non_time_single_contributions = [output(non_time_data).numpy() for output in non_time_single_outputs]
    time_pairwise_contributions = [output(time_data).numpy() for output in time_pairwise_outputs]
    non_time_pairwise_contributions = [output(non_time_data).numpy() for output in non_time_pairwise_outputs]
    return time_single_contributions, non_time_single_contributions, time_pairwise_contributions, non_time_pairwise_contributions

# Feature importance by averaging absolute contributions
def calculate_feature_importance(contributions):
    importance = [np.mean(np.abs(contrib)) for contrib in contributions]
    return importance

# Prediction breakdown
def prediction_breakdown(time_data, non_time_data, sample_index):
    time_single, non_time_single, time_pairwise, non_time_pairwise = calculate_feature_contributions(time_data, non_time_data)
    breakdown = {
        "Time-Series Single Contributions": time_single[sample_index],
        "Non-Time-Series Single Contributions": non_time_single[sample_index],
        "Time-Series Pairwise Contributions": time_pairwise[sample_index],
        "Non-Time-Series Pairwise Contributions": non_time_pairwise[sample_index],
    }
    return breakdown

# Visualizations for interpretability and explainability
def plot_feature_importance(feature_importance, title):
    plt.bar(range(len(feature_importance)), feature_importance)
    plt.title(title)
    plt.xlabel("Feature Index")
    plt.ylabel("Importance")
    plt.show()

# Training the model
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

print("Training Single Effects...")
train_single_effect(X_time_train, X_non_time_train, y_train, optimizer, epochs=5)

print("Training Pairwise Effects...")
train_pairwise_effect(X_time_train, X_non_time_train, y_train, optimizer, epochs=5)

print("Fine-Tuning...")
fine_tune_model(X_time_train, X_non_time_train, y_train, optimizer, epochs=5)

# Calculate feature contributions and importance
time_single, non_time_single, time_pairwise, non_time_pairwise = calculate_feature_contributions(X_time_test, X_non_time_test)

time_single_importance = calculate_feature_importance(time_single)
non_time_single_importance = calculate_feature_importance(non_time_single)
time_pairwise_importance = calculate_feature_importance(time_pairwise)
non_time_pairwise_importance = calculate_feature_importance(non_time_pairwise)

print("Time-Series Single Feature Importance:", time_single_importance)
print("Non-Time-Series Single Feature Importance:", non_time_single_importance)
print("Time-Series Pairwise Feature Importance:", time_pairwise_importance)
print("Non-Time-Series Pairwise Feature Importance:", non_time_pairwise_importance)

# Plot feature importance
plot_feature_importance(time_single_importance, "Time-Series Single Feature Importance")
plot_feature_importance(non_time_single_importance, "Non-Time-Series Single Feature Importance")
plot_feature_importance(time_pairwise_importance, "Time-Series Pairwise Feature Importance")
plot_feature_importance(non_time_pairwise_importance, "Non-Time-Series Pairwise Feature Importance")

# Prediction breakdown for a sample
sample_index = 0
breakdown = prediction_breakdown(X_time_test, X_non_time_test, sample_index)
print(f"Prediction Breakdown for Sample {sample_index}:")
for key, value in breakdown.items():
    print(key, ":", value)



```
